```julia
using JuMP
using Gurobi
```

```julia
time_=8760
F=[0.03*2000*1000,0.03*950*1000,0.03*700*1000]
EA=[175.2252157*1000,85.60445144*1000,61.91153839*1000]
C=[1,1,1]
l=5000
H=[2.4,1.62,2.5]
P=[5,30,30]
```

```julia
using CSV
using DataFrames

df = CSV.read("data.csv", DataFrame)
D_t = df[!, "FR_load_forecast_entsoe_transparency"]
println(size(D_t))
```

```julia
model=Model(Gurobi.Optimizer)
@variable(model,0<=p[1:3,1:time_])
@variable(model,0<=k[1:3])
@variable(model,0<=d[1:time_])
@objective(model,Min,sum((F[g]+EA[g])*k[g]+sum(H[g]*P[g]*p[g,i] for i in 1:time_) for g in 1:3)-sum(l*d[i] for i in 1:time_))
@constraint(model,[g in 1:3,i in 1:time_],p[g,i] <= C[g]*k[g])
@constraint(model,[i in 1:time_],d[i] <= D_t[i])
@constraint(model,price[i in 1:time_],d[i]==sum(p[g,i] for g in 1:3))
```

Question (i) : optimal installed capacity mix

```julia
optimize!(model)
println("Variables value:",JuMP.value.(k))
println("Objective value: ", objective_value(model))
```

```julia
prices=[-JuMP.dual(price[i]) for i in 1:time_]
println(prices)
```
